```python
from elasticsearch import Elasticsearch
import pandas as pd
import json
from pandas import DataFrame
import re
```

```python


es=Elasticsearch('https://127.0.0.1:9200/',
                 verify_certs=False,
                 basic_auth=('elastic','elastic'))

# 定义查询
query = {
     "_source": ["LGA_code", "2021Persons_total"],
     "query": {
    "match_all": {}
    },'size':1000
}

# 执行搜索
response = es.search(index='population2021', body=query)
```

```python
data=response['hits']['hits']

df=pd.DataFrame(data)

df=pd.DataFrame(list(df["_source"]))
```

```python
es=Elasticsearch('https://127.0.0.1:9200/',
                 verify_certs=False,
                 basic_auth=('elastic','elastic'))

# 定义查询
query = {
     "_source": ["geometry", "siteHealthAdvices.averageValue"],
     "query": {
    "match_all": {}
    },'size':1000
}

# 执行搜索
response_env = es.search(index='environmental_data', body=query)
```

```python
print(response_env)
```

```python
data_env=response_env['hits']['hits']

df_env=pd.DataFrame(data_env)

df_env=pd.DataFrame(list(df_env["_source"]))

df_env
```

```python
df_env=df_env.dropna()
df_env
```

```python
df_env['siteHealthAdvices'] = df_env['siteHealthAdvices'].astype(str)
```

```python
df_env['siteHealthAdvices'] = df_env['siteHealthAdvices'].str.findall(r'\d+\.\d+|\d+').apply(' '.join)
df_env['siteHealthAdvices'] = df_env['siteHealthAdvices'].astype(float)
```
